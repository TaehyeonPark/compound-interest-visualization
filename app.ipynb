# Compound Interest Visualization

## Visualize Multiple Graphs

```python
df1 = cpi.compound_growth(principal=100000, annual_rate=0.1,period_per_year=2, years=30)
df2 = cpi.compound_growth(principal=100000, annual_rate=0.1,period_per_year=1, years=30)
df3 = cpi.compound_growth(principal=40000, annual_rate=0.1,period_per_year=2, years=30)
df4 = cpi.compound_growth(principal=1000, annual_rate=0.1,period_per_year=2, years=30)
cpi.plot([df1, df2, df3, df4])
```

## Project A onto B

```python
df_a = cpi.compound_growth(principal=100000, annual_rate=0.1,period_per_year=2, years=30)
df_b = cpi.compound_growth(principal=40000, annual_rate=0.1,period_per_year=2, years=30)
df = cpi.project_asset(df_a, df_b)
cpi.plot([df_a, df_b, df])
```
